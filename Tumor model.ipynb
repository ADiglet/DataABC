### Tumor model


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from scipy import integrate
from scipy.integrate import odeint
```

```python
def tumor(P0,t,C, P, Q, Qp,KDE,lambdaP,K,kQpP,kPQ,kQp,gammaP,gammaQ,gammaQP):
    Pprim = P + Q + Qp
    dC = -KDE * C
    dP = lambdaP * P(1-(Pprim/K))+kQpP*Qp-kPQ * P - gammaP * C * KDE * P
    dQ = kQp*P - gammaP * C * KDE * Q
    dQp = gammaQ * C * KDE * Q - kQpP * Qp - gammaQP * Qp
    
    return Pprim
```
